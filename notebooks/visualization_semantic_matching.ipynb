```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# --- Load your results file ---
df = pd.read_csv("../data/processed/semantic_matching_output/test3/top_matches_with_missing_skills.csv")

# --- Basic stats ---
print("Summary by JD Category:")
print(df.groupby("Top_JD_Categories")["Top_Score"].agg(["mean", "std", "min", "max", "count"]))
print("\nOverall Summary:")
print(df["Top_Score"].describe())

# --- Optional: visualize distribution ---
plt.figure(figsize=(8,5))
df.groupby("Top_JD_Categories")["Top_Score"].plot(kind='kde')
plt.legend(df["Top_JD_Categories"].unique())
plt.title("Similarity Score Density by Job Category")
plt.xlabel("Cosine Similarity")
plt.show()

# --- PCA visualization (embedding space sanity check) ---
# If you saved embeddings in a file, load them; if not, skip this part.
# Example assuming you have CV and JD embeddings in CSVs or arrays:
# cv_emb = np.load("../data/processed/semantic_matching_output/test3/cv_embeddings.npy")
# jd_emb = np.load("../data/processed/semantic_matching_output/test3/jd_embeddings.npy")

# pca = PCA(n_components=2)
# all_emb = np.vstack([cv_emb, jd_emb])
# reduced = pca.fit_transform(all_emb)
# labels = ["CV"] * len(cv_emb) + ["JD"] * len(jd_emb)
# plt.figure(figsize=(7,6))
# plt.scatter(reduced[:,0], reduced[:,1], c=["blue" if l=="CV" else "red" for l in labels], alpha=0.5)
# plt.title("PCA of CV vs JD Embeddings")
# plt.show()
```
