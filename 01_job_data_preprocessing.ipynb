## Job Title Simplification

```python
df['Job Title'].value_counts()
```

```python
unique_titles = df['Job Title'].unique()
print(f'Total unique job titles: {len(unique_titles)}')
print(unique_titles)
```

```python
def simplify_job_title(title):
    title = title.lower()
    if 'machine learning' in title or 'ml engineer' in title:
        return 'Machine Learning Engineer'
    if 'data scientist' in title:
        return 'Data Scientist'
    if 'data analyst' in title:
        return 'Data Analyst'
    else:
        return 'Other'

df['Simplified Job Title'] = df['Job Title'].apply(simplify_job_title)
df['Simplified Job Title'].value_counts()
```

```python
df[df['Simplified Job Title'] == 'Other']['Job Title'].value_counts().head(30)
```

```python
def simplify_job_title(title):
    title = title.lower()
    
    if any(x in title for x in ["machine learning", "ml", "deep learning", "ai", "research engineer"]):
        return "Machine Learning Engineer"
    elif any(x in title for x in ["data scientist", "science", "scientist", "modeler", "r&d", "quantitative", "analytical"]):
        return "Data Scientist"
    elif any(x in title for x in ["data engineer", "platform", "etl", "pipeline", "architecture", "database"]):
        return "Data Engineer"
    elif any(x in title for x in ["data analyst", "analytics", "bi", "business analyst", "consultant", "manager"]):
        return "Data Analyst"
    elif any(x in title for x in ["software", "developer", "programmer"]):
        return "Software Engineer"
    else:
        return "Other"

df['Simplified Job Title'] = df['Job Title'].apply(simplify_job_title)
df['Simplified Job Title'].value_counts()
```

```python
df = df[df['Simplified Job Title'] != 'Other']
```

## Text Preprocessing

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 2]
    return " ".join(tokens)

df["clean_text"] = df["Job Description"].apply(clean_text)
```

```python
df.to_csv("data/glassdoor_jobs_cleaned.csv", index=False)
```
